```python
chars_to_remove = ["+",",","$"]
cols_to_clean = ["Installs","Price"]
for item in chars_to_remove:
    for cols in cols_to_clean:
        data_frame[cols] = data_frame[cols].str.replace(item,"")
```

```python
data_frame["Last Updated"] = pd.to_datetime(data_frame["Last Updated"])
data_frame["Day"] = data_frame["Last Updated"].dt.day
data_frame["Month"] = data_frame["Last Updated"].dt.month
data_frame["Year"] = data_frame["Last Updated"].dt.year
```

```python
numeric_features = [feature for feature in data_frame.columns if data_frame[feature].dtype != 'O']
categorical_features = [feature for feature in data_frame.columns if data_frame[feature].dtype == 'O']
```

```python
plt.figure(figsize=(15, 15))
plt.suptitle('Numerical Features', fontsize=18, fontweight='bold')

for i in range(0, len(numeric_features)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=data_frame[numeric_features[i]],fill=True, color='b')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.suptitle('Categorical Features', fontsize=18, fontweight='bold')
category = ['Type', 'Content Rating']
for i in range(0, len(category)):
    plt.subplot(2, 2, i+1)
    sns.countplot(x=data_frame[category[i]])
    plt.xlabel(category[i])
    plt.tight_layout() 
plt.show()
```

```python
data_frame_category_installs = data_frame.groupby(['Category'])['Installs'].sum().sort_values(ascending = False).reset_index()
```

```python
data_frame_category_installs.Installs = data_frame_category_installs.Installs/1000000000 
```

```python
data_frame2 = data_frame_category_installs.head(10)
plt.figure(figsize = (15,5))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(x = 'Installs' , y = 'Category' , data = data_frame2 )
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('')
ax.set_title("Most Popular Categories in Play Store", size = 20)
plt.show()
```

```python
data_frame_app_category = data_frame.groupby(['Category' ,'App'])['Installs'].sum().reset_index()
data_frame_app_category = data_frame_app_category.sort_values('Installs', ascending = False)
apps = ['GAME', 'COMMUNICATION', 'PRODUCTIVITY', 'SOCIAL' ,'TOOLS']
sns.set_context("poster")
sns.set_style("darkgrid")

plt.figure(figsize=(40,30))

for i,app in enumerate(apps):
    data_frame2 = data_frame_app_category[data_frame_app_category.Category == app]

    
    data_frame_top_5 = data_frame2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data= data_frame_top_5,x= 'Installs' ,y='App' )
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size = 20)
    
plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```
